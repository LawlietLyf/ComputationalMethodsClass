# ApproxFun.jl Examples

Here's some code to show what ApproxFun.jl is and what it can do.

```julia
#Set up the Julia environment to use standard packages using Project and Manifest in file directory
import Pkg; 
Pkg.activate(".")
Pkg.instantiate()
```

```julia
using LinearAlgebra, SpecialFunctions, Plots, ApproxFun
```

```julia
gr(fmt=:png); # setting for easier display in jupyter notebooks
```

```julia
x = Fun(identity,0..10) #Define a function, which is just x=x
```

```julia
f = sin(x^2)
```

```julia
g = cos(x)
```

```julia
h = f + g^2
```

```julia
r = roots(h)
```

```julia
rp = roots(h') #' here means derivative!
```

```julia
plot(h; label="f + g^2")
scatter!(r,h.(r); label="roots")
scatter!(rp,h.(rp),label="extrema")
```
